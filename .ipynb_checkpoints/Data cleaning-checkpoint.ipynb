# Proyecto limpieza de datos

## Objetivos


- Obtener los datos limpios donde todas las personas sean residentes de USA
- Analizar datos para saber el salario promedio anual en USA
- Analizar datos para saber cuales son las industrias que pagan mas en promedio
- Analizar datos para saber en promedio que estados de USA pagan mas
- Analizar datos para saber cuales son los trabajos que otorgan mejores bonos

## Limpieza de datos

Importando librerias necesarias para la limpieza de datos

```python
import pandas as pd
import numpy as np
```

Base de datos a limpiar

```python
df = pd.read_csv("Data\Ask A Manager Salary Survey 2021 (Responses) - Form Responses 1.csv", index_col='Timestamp', parse_dates=['Timestamp'])
df.head(5)
```

Cambiando el nombre de las columnas para mayor comodidad

```python
df.columns = ['edad', 'industria', 'trabajo', 'contexto trabajo', 'salario anual', 'bonos', 'moneda', 'otras monedas', 'contexto entrada', 'pais de trabajo', 'estado', 'ciudad', 'experiencia en total', 'experiencia en la industria', 'educacion', 'genero', 'etnia']
df.head(5)
```

Eliminando las columnas innecesarias

```python
df = df.drop(columns=['contexto trabajo', 'contexto entrada', 'otras monedas', 'ciudad', 'experiencia en total', 'experiencia en la industria', 'educacion', 'genero', 'etnia'])
df.head(5)
```

Remplazando las celdas nulas en la columna bonos por ceros y convirtiendo la columna en numeros enteros

```python
df['bonos'] = df['bonos'].replace(np.nan, 0).astype(int)
df.dtypes
```

Eliminando las comas de los dijitos y convirtiendo la columna en numeros enteros

```python
df['salario anual'] = df['salario anual'].str.replace(",", "").astype(int)
df.dtypes
```

Eliminando los espacios de las cadenas de texto para facilitar la busqueda de paises

```python
df['pais de trabajo'] = df['pais de trabajo'].str.replace(" ", "")
df.head(5)
```

Usando regex para encontrar todas aquellas personas que usaron distintos tipos de nomenclatura para el pais USA 

ejemplo (US, United states, U.S.A )

y reemplazandolo por la palabra USA para asi agrupar todas las personas que trabajan en estados unidos en una nomenclatura para todos

```python
df['pais de trabajo'] = df['pais de trabajo'].str.replace("^U\.[^kKaA].+", "USA", regex=True)
df['pais de trabajo'] = df['pais de trabajo'].str.replace("^Un....[^kKaA].+", "USA", regex=True)
df['pais de trabajo'] = df['pais de trabajo'].str.replace("^U[sS]", "USA", regex=True)
df['pais de trabajo'] = df['pais de trabajo'].str.replace("^TheU.+", "USA", regex=True)
df['pais de trabajo'] = df['pais de trabajo'].str.replace("^USA.+", "USA", regex=True)
```

Obteniendo los datos limpios y listos para analizar y sacar informacion que ayude a la toma de decisiones

```python
clean_data = df[df['pais de trabajo'].str.contains("^USA")]
clean_data.head(5)
```

# Analizando datos

Obteniendo el salario promedio anual en USA

```python
promedio_salario_anual_USA = clean_data.groupby('pais de trabajo')['salario anual'].mean().tolist()    
promedio_salario_anual_USA = '${:,.2f}'.format(promedio_salario_anual_USA[0])
print(f'El salario promedio anual en los estados unidos segun los datos dados es de {promedio_salario_anual_USA}')
```

Obteniendo las industrias que pagan mas en promedio

```python
industrias_mejor_pagadas = clean_data.groupby('industria')['salario anual'].mean().reset_index().sort_values(by='salario anual', ascending=False).head(5)
industrias_mejor_pagadas['salario anual'] = industrias_mejor_pagadas['salario anual'].map('${:,.2f}'.format)
industrias_mejor_pagadas.columns = ['Industrias', 'Salario Promedio Anual']
industrias_mejor_pagadas
```

Obteniendo los estados de USA que pagan mas en promedio

```python
estados_mejor_pagan = clean_data.groupby('estado')['salario anual'].mean().reset_index().sort_values(by='salario anual', ascending=False).head(5)
estados_mejor_pagan['salario anual'] = estados_mejor_pagan['salario anual'].map('${:,.2f}'.format)
estados_mejor_pagan.columns = ['Estados', 'Salario Promedio Anual']
estados_mejor_pagan
```

Obteniendo las industrias que otorgan mejores bonos

```python
trabajos_mejores_bonos = clean_data.groupby('trabajo')['bonos'].mean().reset_index().sort_values(by='bonos', ascending=False).head(5)
trabajos_mejores_bonos['bonos'] = trabajos_mejores_bonos['bonos'].map('${:,.2f}'.format)
trabajos_mejores_bonos.columns = ['Trabajo', 'Bonos']
trabajos_mejores_bonos
```
